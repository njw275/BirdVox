```python
import numpy as np
import Paths as p
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt
import DetectionEvalSourceCode as det

duration_length = 605

file = p.get_data() + "unit" + str(unit).zfill(2) + "_S_" + str(start_time) + "_E_" + str(start_time+605) + ".wav"

y, sr = librosa.load(file, duration=duration_length)
S = np.abs(librosa.stft(y))
D = librosa.power_to_db(S**2,ref=np.max)
    
librosa.display.specshow(D,  x_axis='time', y_axis='log')
plt.colorbar(format='%+2.0f dB')
plt.title('Power spectrogram')
plt.show()
```
